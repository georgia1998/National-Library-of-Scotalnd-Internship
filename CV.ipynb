While, I have no work experience in programming I do have experience i hospitality which as allowed me to gain confidence and comunication skills. Here is an overview of my work history:

```python
dates = ['06-2016','09-2016','07-2017', '08-2018', '06-2019', '09-2019','10-2019','04-2020']
names = ['First Job','Left To Start Uni','Started work at "Five Guys"','Left to Study Abroad',
         'Began Work at "Birdhouse Tearoom"', 'Left to Return to Uni',
         'Started at "Waterstones"','Finished Uni']
levels = np.tile([-5, 5, -3, 3, -6, 6],
                 int(np.ceil(len(dates)/6)))[:len(dates)]

fig, ax = plt.subplots(figsize=(8.8, 4), constrained_layout=True)
ax.set(title="Job History")

markerline, stemline, baseline = ax.stem(dates, levels,
                                         linefmt="g-", basefmt="b-",
                                         use_line_collection=True)

plt.setp(markerline, mec="k", mfc="w", zorder=3)

markerline.set_ydata(np.zeros(len(dates)))


vert = np.array(['top', 'bottom'])[(levels > 0).astype(int)]
for d, l, r, va in zip(dates, levels, names, vert):
    ax.annotate(r, xy=(d, l), xytext=(-3, np.sign(l)*3),
                textcoords="offset points", va=va, ha="right")


plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

ax.get_yaxis().set_visible(False)
for spine in ["left", "top", "right"]:
    ax.spines[spine].set_visible(False)

ax.margins(y=0.2)
plt.show()
```
